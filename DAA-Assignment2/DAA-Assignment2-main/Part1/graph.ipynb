```python
import networkx as nx
import matplotlib.pyplot as plt
sr = sink = 0
with open('input.txt', 'r') as f:
    vertexs = int(f.readline())
    sr = int(f.readline())
    sink = int(f.readline())
    edges = int(f.readline())
    G = nx.DiGraph()
    for i in range(0,edges):
        u,v,c= [int(x) for x in next(f).split()]
        G.add_edge(u, v, capacity=c, flow=0)
color_map = []
for node in G:
    if node == sr:
        color_map.append('yellow')
    elif node == sink:
        color_map.append('red')
    else: 
        color_map.append('lightblue')
# Visualize graph
plt.figure(figsize=(10,10))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color=color_map)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{d['flow']}/{d['capacity']}" for u, v, d in G.edges(data=True)},font_size=12)
nx.draw_networkx_labels(G, pos)
plt.axis('off')
plt.show()
```

```python
with open('output1.txt', 'r') as f:
    flows = f.readline()
    no_of_flows = int(f.readline())
    for i in range(0,no_of_flows):
        u,v,nf= [int(x) for x in next(f).split()]
        G[u][v]['flow'] = nf
plt.figure(figsize=(10,10))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color=color_map)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{d['flow']}/{d['capacity']}" for u, v, d in G.edges(data=True)},font_size=12)
nx.draw_networkx_labels(G, pos)
plt.axis('off')
plt.show()
```

```python
with open('output3.txt','r') as f:
    A = []
    B = [] 
    flows = f.readline()
    no_of_flows = int(f.readline())
    G = nx.Graph()
    test_list=[]
    for i in range(0,no_of_flows):
        u,v,nf= [int(x) for x in next(f).split()]
        if nf != 0:
            A.append(u)
            B.append(v)
            test_list.append((u,v))
G.add_nodes_from(A, bipartite=0) # nodes in A belong to set 0
G.add_nodes_from(B, bipartite=1) # nodes in B belong to set 1
# plt.figure(figsize=(10,10))
G.add_edges_from(test_list)

pos = nx.bipartite_layout(G,A)  # Generate layout for bipartite graph
nx.draw_networkx_nodes(G, pos, nodelist=A, node_color='g')
nx.draw_networkx_nodes(G, pos, nodelist=B, node_color='y')
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)

plt.axis('off')
plt.show()
```
